## Initialize the global variables

```python
# GeumGang
code_dict = [{'3011685' : '세종시(미호교)'},
             {'3011695' : '세종시(월산교)'},
             {'3011675' : '세종시(상조천교)'},
             {'3011615' : '진천군(신정교)'},
             {'3011617' : '진천군(금곡교)'},
             {'3011620' : '진천군(가산교)'},
             {'3011623' : '진천군(인산리)'},
             {'3011630' : '청주시(여암교)'},
             {'3011635' : '청주시(팔결교)'},
             {'3011665' : '청주시(미호천교)'}]
```

```python
RIVER_NAME = '미호천'
TIME_RANGE = '1D'
DOC_TYPE = '.xml'
API_URL = 'http://api.hrfco.go.kr/'+apiKey+'/waterlevel/list/'
START_DATE = '0101'
END_DATE='1231'
START_YEAR = 2010
END_YEAR = 2023
YEARS_LIST = [str(x) for x in range(START_YEAR,END_YEAR,1)]
```

```python
for r in code_dict:
    
    # River's code
    river = list(r.keys())[0]
    # River's name
    river_name = list(r.values())[0]

    tmp = pd.DataFrame()
    print(river_name+' set starts!')
    
    for y in YEARS_LIST:
        start_dates = y+START_DATE
        end_dates = y+END_DATE
        
        # Request API
        url = os.path.join(API_URL,TIME_RANGE,river,start_dates,end_dates+DOC_TYPE)
        res = requests.get(url)
        
        # XML to DICT
        xmlToDict = xmltodict.parse(res.text) # return collections.OrderedDict
        
        # DICT to JSON
        df_dict = json.loads(json.dumps(xmlToDict)) # return dict
        
        df= df_dict['entities']['content']['Waterlevel']
        data_df = pd.DataFrame(df)
        # Dates, wl, fw
        del data_df['wlobscd']
        tmp = pd.concat([tmp, data_df],axis=0)
    
    # wl : 수위, fw : 유량, Dates : 시간
    tmp.columns=['유량(m3/s)', '수위(m)', '시간']
    tmp['시간'] = pd.to_datetime(tmp['시간'])
    
    # Sorting by the dates
    data_res = tmp.sort_values(by='시간')
    result = data_res.reset_index(drop=True)
    
    # Change the column's order
    result = result[['시간', '수위(m)', '유량(m3/s)']]
    
    # Save the files (csv)
    result.to_csv('./'+RIVER_NAME+'_'+river_name+'_'+str(START_YEAR)+str(START_DATE)+'_'+end_dates+'.csv',encoding='utf-8',index=False)
    print('./'+RIVER_NAME+'_'+river_name+'_'+str(START_YEAR)+str(START_DATE)+'_'+end_dates+'.csv set saved!') 
```
